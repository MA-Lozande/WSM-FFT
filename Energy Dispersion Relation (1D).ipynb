```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
L=1000
k=np.fft.fftfreq(L)*2*np.pi
k=np.fft.fftshift(k)
t=2
```

```python
E1D_1plus=(np.sin(k))-t*np.sin(k)
E1D_1minus=-(np.sin(k))-t*np.sin(k)

plt.figure(figsize=(6,4))

plt.plot(k,E1D_1plus,label=r'$E_{1+}^{1D}$', color='red')
plt.plot(k,E1D_1minus,label=r'$E_{1-}^{1D}$', color='blue')

plt.axhline(y=0, color='black', label='Fermi Energy $E_F$', linestyle='--')

plt.xlabel(r'$k$')
plt.ylabel(r'$E$')
plt.grid(True)
plt.savefig('Energy_dispersion_1', dpi=600, bbox_inches='tight')
plt.show()
```

```python
E1D_2plus2=np.sqrt(np.sin(k)**2 + (1-np.cos(k))**2)-t*np.sin(k)
E1D_2minus2=-np.sqrt(np.sin(k)**2 + (1-np.cos(k))**2)-t*np.sin(k)
E1D_2plus=2*np.sin(k/2) - t*np.sin(k)
E1D_2minus=-2*np.sin(k/2) - t*np.sin(k)

plt.figure(figsize=(6,4))

plt.plot(k,E1D_2plus2,label=r'$E_{2+2}^{1D}$', color='green')
plt.plot(k,E1D_2minus2,label=r'$E_{2-2}^{1D}$', color='yellow')
plt.plot(k,E1D_2plus,label=r'$E_{2+}^{1D}$', color='red', linestyle='--')
plt.plot(k,E1D_2minus,label=r'$E_{2-}^{1D}$', color='blue', linestyle='--')

plt.axhline(y=0, color='black', label='Fermi Energy $E_F$', linestyle='--')

plt.xlabel(r'$k$')
plt.ylabel(r'$E$')
plt.grid(True)
plt.savefig('Energy_dispersion_2', dpi=600, bbox_inches='tight')
plt.show()
```
